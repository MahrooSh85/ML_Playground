Here I show how to work with scikilearn dataset. For example, I choose breast_cancer dataset. 

```python
import numpy as np
import pandas as pd
```

## Load the dataset

```python
from sklearn.datasets import load_breast_cancer

breast_cancer = load_breast_cancer()
```

## Print the description of dataset 

```python
print(breast_cancer.DESCR)
```

## Find the type of dataset

```python
type(breast_cancer)
```

The object returned by **load_breast_cancer()** is a scikit-learn Bunch object, which is similar to a dictionary. So we use the **.keys()** to find the keys.

```python
breast_cancer.keys()
```

```python
breast_cancer.feature_names
```

## How to make a Pandas DataFrame

```python
df = pd.DataFrame(data=breast_cancer.data, columns=breast_cancer.feature_names)
df['target'] = breast_cancer.target
```

```python
print(df.shape)
df.head()
```

Find the diversity of target values

```python
df.target.value_counts(ascending=True)
```

Ignore the target columns

```python
df.iloc[:,:-1]
```
